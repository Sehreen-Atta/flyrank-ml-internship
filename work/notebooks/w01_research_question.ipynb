## 1. My lane (or freestyle) and why

**Lane 3: Structured Content Archetype Clustering**

I'm grouping pages by how they *behave* in search and engagement metrics — not by topic or article text — to find recurring performance archetypes across the inventory. ## 2. The question: decision, action, cost of a wrong call

**Question:** What performance archetypes exist across the content inventory, and what action fits each one?

| Piece | Answer |
|---|---|
| **Unit of analysis** | One content item (one row per page) |
| **Output** | Cluster assignments + a profile per cluster (typical impressions, CTR, position, freshness, trend direction) |
| **Decision improved** | Which *type* of page this is, so the reviewer picks the right playbook instead of starting from scratch |
| **Who acts** | A content reviewer / SEO analyst who applies archetype-specific actions: protect, improve, rewrite, merge, prune, or monitor |
| **Cost of a wrong call** | Mis-bucketing — e.g. a real "hidden gem" labeled "weak/no-demand" never gets attention, or a genuinely stale page grouped with healthy ones gets skipped |
| **Why data/ML helps** | Manually eyeballing tens of thousands of pages doesn't scale; clustering surfaces natural groupings from the numbers themselves rather than a human guessing categories upfront |

This is **not** "train a classifier to predict decline." It's unsupervised grouping: no single ranked score, but a reusable map of how pages behave in the metrics I choose.

```python
frame = {
    "question": "What performance archetypes exist, and what action fits each?",
    "unit": "one content item (one row per page)",
    "task_type": "clustering (unsupervised)",
    "output": "cluster labels + metric profiles + action mapping",
    "actor": "content reviewer applying archetype playbooks",
}
for k, v in frame.items():
    print(f"{k}: {v}")
```

## 3. Quick look at the data (2-3 real numbers)

The starter CSV has **30,000 pages across 32 clients** — enough scale that manual archetype labeling isn't realistic. Three signals that clustering has real variation to work with (numbers from the code cell below):

1. **Trend direction is spread, not flat:** 16,262 pages are `down`, 5,962 `stable`, 4,388 `up`, plus `new` and `flat` — several distinct movement patterns to separate.
2. **Search visibility varies wildly:** `impressions_90d` median is 731 but the max is 517,715 (25th percentile 81, 75th percentile 3,615) — high-visibility and low-visibility archetypes can coexist.
3. **Position mix is diverse:** 11,814 pages sit on page 1, 7,304 in "striking distance," 7,242 on pages 3–5, and 1,319 deep — position tier alone creates meaningfully different performance contexts to cluster on.

```python
from pathlib import Path

import pandas as pd

CSV = Path("../../data/raw/content_refresh_anonymized.csv")
df = pd.read_csv(CSV)

print(f"rows: {len(df):,}  |  clients: {df['client_id'].nunique()}")
print()
print("trend_direction:")
print(df["trend_direction"].value_counts().to_string())
print()
print("impressions_90d (spread):")
print(df["impressions_90d"].describe().round(1).to_string())
print()
print("position_tier:")
print(df["position_tier"].value_counts().to_string())
```

## 4. Careful words: what I can and can't claim

Clusters describe **patterns in the metrics I chose** — impressions, CTR, position, freshness, trend — not the article's actual topic or writing quality. Cluster membership is **observational and directional**: it suggests which playbook might fit, but it is not proof of causation (grouping a page with "declining" pages does not prove a refresh will fix it). There is **no article text** in this dataset, so this is metric-based structured grouping — I will never call it semantic clustering or claim I discovered what a page is "about."

```python
CAN_CLAIM = [
    "observed metric patterns group into separable archetypes",
    "directional action mapping (protect / improve / rewrite / merge / prune / monitor)",
    "decision-support for reviewers triaging at scale",
]
CANNOT_CLAIM = [
    "semantic or topic-based clustering (no text in the data)",
    "causal proof that an action will recover traffic",
    "Google algorithm factors or ranking guarantees",
]
print("Can claim:", CAN_CLAIM)
print("Cannot claim:", CANNOT_CLAIM)
```
